```python
import pandas as pd
import numpy as np
```

```python
dados_totais = pd.read_csv('./data/datasus_SIH.csv', encoding='cp850', sep=';', engine='python')
dados_totais
```

```python
print(type(dados_totais))
print('---------')
print(dados_totais.shape)
print('---------')
print(dados_totais.dtypes)
```

```python
# Renomear coluna do município
dados_totais = dados_totais.rename(columns={'Municipio': 'municipio'})

dados_totais
```

```python
# tirar as linhas que contenham municipio ignorado.
dados_totais = dados_totais[dados_totais['municipio'].apply(lambda x: x is not None and 'MUNICIPIO IGNORADO' not in x)]
dados_totais
```

```python
#tirei tbm as tres ultimas linhas
dados_totais = dados_totais.iloc[:-3]
dados_totais
```

```python
# Separar o código do IBGE em uma coluna separada para usá-lo como key
dados_totais['cod_ibge'] = dados_totais['municipio'].str.split(" ", n=1, expand = True)[0]
dados_totais['municipio'] = dados_totais['municipio'].str.split(" ", n=1, expand = True)[1]
dados_totais.head(10)
```

```python
dados_totais['Cod_UF'] = dados_totais['cod_ibge'].str[:2]
dados_totais
```

```python
dados_totais.info()
```

```python
# Criando UF e região
CondUF =  [(dados_totais['Cod_UF'] == '11'),
         (dados_totais["Cod_UF"] == '12'),
         (dados_totais["Cod_UF"] == '13'),
         (dados_totais["Cod_UF"] == '14'),
         (dados_totais["Cod_UF"] == '15'),
         (dados_totais["Cod_UF"] == '16'),
         (dados_totais["Cod_UF"] == '17'),
         (dados_totais["Cod_UF"] == '21'),
         (dados_totais["Cod_UF"] == '22'),
         (dados_totais["Cod_UF"] == '23'),
         (dados_totais["Cod_UF"] == '24'),
         (dados_totais["Cod_UF"] == '25'),
         (dados_totais["Cod_UF"] == '26'),
         (dados_totais["Cod_UF"] == '27'),
         (dados_totais["Cod_UF"] == '28'),
         (dados_totais["Cod_UF"] == '29'),
         (dados_totais["Cod_UF"] == '31'),
         (dados_totais["Cod_UF"] == '32'),
         (dados_totais["Cod_UF"] == '33'),
         (dados_totais["Cod_UF"] == '35'),
         (dados_totais["Cod_UF"] == '41'),
         (dados_totais["Cod_UF"] == '42'),
         (dados_totais["Cod_UF"] == '43'),
         (dados_totais["Cod_UF"] == '50'),
         (dados_totais["Cod_UF"] == '51'),
         (dados_totais["Cod_UF"] == '52'),
         (dados_totais["Cod_UF"] == '53')
         ]
          
UF_nomes = ['RO','AC','AM','RR','PA','AP','TO','MA','PI','CE','RN','PB','PE','AL','SE','BA','MG','ES','RJ','SP','PR','SC','RS','MS','MT','GO','DF']
dados_totais["UF"] = np.select(CondUF, UF_nomes, default="Error")

CondReg =  [(dados_totais['Cod_UF'] == '11') | (dados_totais["Cod_UF"] == '12') | (dados_totais["Cod_UF"] == '13') | (dados_totais["Cod_UF"] == '14') | (dados_totais["Cod_UF"] == '15') | (dados_totais["Cod_UF"] == '16') | (dados_totais["Cod_UF"] == '17'),
            (dados_totais["Cod_UF"] == '21') | (dados_totais["Cod_UF"] == '22') | (dados_totais["Cod_UF"] == '23') | (dados_totais["Cod_UF"] == '24') | (dados_totais["Cod_UF"] == '25') | (dados_totais["Cod_UF"] == '26') | (dados_totais["Cod_UF"] == '27') | (dados_totais["Cod_UF"] == '28') | (dados_totais["Cod_UF"] == '29'),
            (dados_totais["Cod_UF"] == '31') | (dados_totais["Cod_UF"] == '32') | (dados_totais["Cod_UF"] == '33') | (dados_totais["Cod_UF"] == '35'),
            (dados_totais["Cod_UF"] == '41') | (dados_totais["Cod_UF"] == '42') | (dados_totais["Cod_UF"] == '43'),
            (dados_totais["Cod_UF"] == '50') | (dados_totais["Cod_UF"] == '51') | (dados_totais["Cod_UF"] == '52') | (dados_totais["Cod_UF"] == '53')
           ]
            
Regiao_nomes = ['Norte','Nordeste','Sudeste','Sul','Centro-Oeste']
dados_totais["regiao"] = np.select(CondReg, Regiao_nomes, default="Error")
dados_totais
```

# dados de valores dos gastos

```python
#criando um df so com os dados de valores
colunas = ['municipio', 'valor_2013', 'valor_2014', 'valor_2015', 'valor_2016', 'valor_2017', 'valor_2018', 'valor_2019', 'valor_2020', 'valor_2021', 'valor_2022', 'cod_ibge', 'Cod_UF', 'UF', 'regiao' ]
dados_valor=dados_totais.loc[:,colunas]
dados_valor
```

```python
# Removendo o prefixo "valor_" do nome das colunas de datas
# removendo ano de 2013

dados_valor = dados_valor.rename(columns=lambda x: x.replace('valor_', ''), inplace=False)

dados_valor = dados_valor.drop(columns=['2013'])
dados_valor
```

```python
# tranformando a base no formato longo: colunas de anos virarão linhas
# renomeando colunas

dados_valor = pd.melt(dados_valor, id_vars=['municipio', 'cod_ibge', 'Cod_UF', 'UF', 'regiao'], var_name='ano', value_name='gastos')

dados_valor = dados_valor.rename(columns={'Cod_UF': 'cod_uf', 'UF': 'uf'})
dados_valor
```

```python
# limpeza das colunas strings

import unicodedata

def preprocess_columns_str(df, column_names):
    for col in column_names:
        df[col] = df[col].apply(lambda x: unicodedata.normalize("NFD", x))
        df[col] = df[col].apply(lambda x: x.encode("ascii", "ignore").decode("utf-8"))
        df[col] = df[col].apply(lambda x: x.upper())
        df[col] = df[col].apply(lambda x: x.strip())
    return df

# Lista de colunas a serem pré-processadas
colunas_a_preprocessar = ['municipio', 'regiao']

# Aplicar a função para pré-processamento das colunas
dados_valor = preprocess_columns_str(dados_valor, colunas_a_preprocessar)

dados_valor
```

```python
# salvando base tratada

dados_valor.to_excel('./dados_tratados/gastos_hospitalares.xlsx', index=False)  #salvar em excel
```

```python
# transformando codigo ibge em string, renomeando coluna população
# retirando o sétimo digito de cod_6

pop22['COD_6'] = pop22['COD_6'].astype(int).astype(str)

pop22 = pop22.rename(columns={'POPULAÇÃO': '2022'})

pop22['cod_ibge'] = pop22['COD_6'].str.slice(stop=-1)
pop22
```

### demais pops

```python
# abrindo base e renomeando coluna municipio
#retirando as 4 ultimas linhas (fonte dos dados e notas explicativas)
# deletando ano de 2013

dados_pop = pd.read_csv('./data/ibge_cnv_popsvs.csv', encoding='cp850', sep=';', engine='python')

dados_pop = dados_pop.rename(columns={'MunicÝpio': 'municipio'})
dados_pop = dados_pop.iloc[:-4]

dados_pop = dados_pop.drop(columns='2013')
dados_pop
```

```python
# Separar código do IBGE e nome do município em duas colunas

dados_pop['cod_ibge'] = dados_pop['municipio'].str.extract('(\d+)')
dados_pop['municipio'] = dados_pop['municipio'].str.extract('([A-Za-z\s\']+)')

dados_pop
```

### mergindo pops com pop22

```python
# juntando bases de população

merged_pops = pd.merge(dados_pop, pop22[['cod_ibge', '2022']],
                     on=['cod_ibge'], how='left')

merged_pops
```

```python
# formato longo da base, anos nas colunas se transformarão em linhas

merged_pops_longo = pd.melt(merged_pops, id_vars=['cod_ibge', 'municipio'],
                     var_name='ano', value_name='populacao')

merged_pops_longo

merged_pops_longo.to_excel('./dados_tratados/populacao.xlsx', index=False)  #salvar em excel
```

# JUNTANDO TODAS AS BASES PARA O DATASET FINAL

```python
emendas = pd.read_excel('./dados_tratados/emendas_to_join.xlsx', engine='openpyxl')

emendas['ano'] = emendas['ano'].astype(str)
emendas.info()
```

```python
# juntando todas as bases numa só

merge = pd.merge(dados_valor, dados_mortalidade[['cod_ibge', 'ano', 'mortalidade']],
                     on=['cod_ibge', 'ano'], how='left')

merge = pd.merge(merge, dados_AIH[['cod_ibge', 'ano', 'aih_valor']],
                     on=['cod_ibge', 'ano'], how='left')

merge = pd.merge(merge, dados_dias[['cod_ibge', 'ano', 'dias']],
                     on=['cod_ibge', 'ano'], how='left')


merge = pd.merge(merge, merged_pops_longo[['cod_ibge', 'ano', 'populacao']],
                     on=['cod_ibge', 'ano'], how='left')


merge = pd.merge(merge, emendas, on=['ano', 'uf'], how='left')

merge['gastos_percapita'] = merge['gastos'] / merge['populacao']

merge.to_excel('./dados_tratados/base_completa.xlsx', index=False)  #salvar em excel

merge
```
