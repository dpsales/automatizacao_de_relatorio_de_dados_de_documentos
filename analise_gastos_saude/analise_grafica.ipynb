# Script de criação das análises visuais do projeto

```python
import pandas as pd
import plotly.express as px

df = pd.read_excel('variaveis_uf_ano.xlsx', engine='openpyxl')

df.head()
```

```python
df.info()
```

### Totais por ano em gráfico de barra

```python
# gráfico de barras para todas as variáveis


fig_barr_total = px.bar(
    df, x='ano', y='gastos', title='Gastos Hospitalares', template='plotly_dark').update_layout(
        title_x=0.5).update_xaxes(
            title = 'Ano',showgrid=False).update_yaxes(
                visible= False, showticklabels=False, showgrid=False)
fig_barr_total.show()
```

### Totais por ano em gráfico scatter

```python
# gráfico scatter para todas as variáveis


fig_scatter_total = px.scatter(
    df, x='ano', y='gastos', title='Gastos Hospitalares', template='plotly_dark').update_layout(
        title_x=0.5).update_xaxes(
            title = 'Ano',showgrid=False).update_yaxes(
                visible= False, showticklabels=False, showgrid=False)
fig_scatter_total.show()
```

### comparações entre as ufs e ano em gráfico de barras

```python
#### comparação uf e ano todas as variáveis em gráfico de barras
fig_barr_uf_ano = px.bar(
    df, x='uf', y='valor_emendas', color='ano', template='plotly_dark').update_layout(
        barmode='relative', title='Evolução dos Gastos Hospitalares - Unidade da Federação', title_x=0.5).update_layout(
            legend_title_text='UF').update_xaxes(
                title = 'Estados', showgrid=False).update_yaxes(
                    visible= False, showticklabels=False, showgrid=False)
fig_barr_uf_ano.show()
```

```python
# coloquei esse no dashboard

df_sorted = df.sort_values(by='gastos', ascending=True)
# Criar o gráfico de barras horizontais
fig_barr_ano_uf_horizontal = px.bar(
    df_sorted, x='gastos', y='uf', color='ano', template='plotly_dark').update_layout(
        barmode='relative', title='Evolução dos Gastos Hospitalares - Unidade da Federação', title_x=0.5).update_layout(
            legend_title_text='Ano').update_xaxes(
                title='Gastos Hospitalares', showgrid=False).update_yaxes(
                    title='Estados', showgrid=False)

fig_barr_ano_uf_horizontal.show()

```

### comparações entre anos e ufs em gráficos de barra

```python

df_sorted = df.sort_values(by='gastos', ascending=True)

# Criar o gráfico de barras
fig_barr_ano_uf = px.bar(
    df_sorted, x='ano', y='gastos', color='uf', template='plotly_dark').update_layout(
        barmode='relative', title='Evolução dos Gastos Hospitalares - Unidade da Federação', title_x=0.5).update_layout(
            legend_title_text='UF').update_xaxes(
                title='Estados', showgrid=False).update_yaxes(
                    visible=False, showticklabels=False, showgrid=False)

fig_barr_ano_uf.show()
```

```python


# Criar o gráfico de barras
fig_barr_ano_uf = px.line(
    df, x='ano', y='gastos', color='uf', template='plotly_dark').update_layout(
        barmode='relative', title='Evolução dos Gastos Hospitalares - Unidade da Federação', title_x=0.5).update_layout(
            legend_title_text='UF').update_xaxes(
                title='Estados', showgrid=False).update_yaxes(
                    visible=False, showticklabels=False, showgrid=False)

fig_barr_ano_uf.show()
```

### análise da relação entre mortalidade e gastos ou gastos per capita

```python
# Criando um gráfico para gastos e gastos_percapita com mortalidade


fig_scatter_mortalidade = px.scatter(
    df, x='mortalidade', y='gastos',title='Gastos Hospitalares X Mortalidade', template='plotly_dark').update_layout(
        title_x=0.5).update_xaxes(
            title='Mortalidade', showgrid=False).update_yaxes(
                title='Gastos Hospitalares', showgrid=False)

fig_scatter_mortalidade.show()
```

### gastos e mortalidade, visualizando anos distintos

```python

df_2022 = df[df['ano'] == 2022].copy()  # Use .copy() para evitar o aviso

# Criar uma nova coluna 'text' contendo o nome da UF
df_2022['text'] = df_2022['uf']

# Criar o gráfico de dispersão apenas para o ano de 2022
fig_scatter_mortalidade = px.scatter(
    df_2022, x='mortalidade', y='gastos', title='Gastos Hospitalares X Mortalidade (2022)', text='text').update_layout(
        title_x=0.5).update_xaxes(
            title='Mortalidade', showgrid=False).update_yaxes(
                title='Gastos Hospitalares', showgrid=False)

fig_scatter_mortalidade.show()

```

### análise da relação entre valor das emendas parlamentares e gastos ou gastos per capita

```python
# Criando um gráfico para gastos e gastos_percapita com mortalidade


fig_scatter_mortalidade = px.scatter(
    df, x='valor_emendas', y='gastos_percapita',title='Gastos Hospitalares X Emendas Parlamentares', template='plotly_dark').update_layout(
        title_x=0.5).update_xaxes(
            title='Emendas Parlamentares', showgrid=False).update_yaxes(
                title='Gastos Hospitalares', showgrid=False)

fig_scatter_mortalidade.show()
```
